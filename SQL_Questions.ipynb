# Question 20 (Interview Level ⭐⭐⭐)

## Scenario

The finance team wants to analyze **high-value delivered orders**.

They only want to consider:

- Orders with the status **Delivered**

After grouping, they want to know **which payment methods have generated more than ₹30,000** from these delivered orders.

Write a SQL query to display:

- `payment_method`
- `total_revenue`

---

## Requirements

- Consider only **Delivered** orders.
- Group the records by `payment_method`.
- Calculate the total revenue.
- Display only payment methods whose revenue is **greater than ₹30,000**.
- Sort the result by **total_revenue** in descending order.

---

## Expected Concepts

- `WHERE`
- `GROUP BY`
- `SUM()`
- `HAVING`
- `ORDER BY`

---

## Sample Output

| payment_method | total_revenue |
|----------------|--------------:|
| UPI | 48,520.00 |
| Credit Card | 44,810.00 |
| Wallet | 38,930.00 |
| Debit Card | 33,760.00 |

> **Note:** The above values are sample outputs. Your actual results may vary depending on your dataset.

---

## Hint

Think about the SQL execution order:

1. Filter individual rows.
2. Group the remaining rows.
3. Calculate the aggregate.
4. Filter the groups.
5. Sort the result.

---

**Difficulty Level:** ⭐⭐⭐ Interview Level
