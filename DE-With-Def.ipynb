# Jupyter SQL Magic Commands

Jupyter Notebook provides **SQL Magic Commands** through the `ipython-sql` extension. These commands allow you to execute SQL queries directly inside notebook cells.

## Load the SQL Extension

```python
%load_ext sql
```

- Loads the **ipython-sql** extension into the current Jupyter Notebook session.
- This command only needs to be executed **once** after opening the notebook.

## Connect to a MySQL Database

```python
%sql mysql+mysqlconnector://root:123456@localhost/test
```

This command establishes a connection to the MySQL database.

### Connection String Breakdown

| Part | Description |
|------|-------------|
| `mysql+mysqlconnector` | Database dialect and Python driver used to connect to MySQL. |
| `root` | MySQL username. |
| `123456` | Password for the MySQL user. |
| `localhost` | Database server address. |
| `test` | Database name to connect to. |

Once the connection is successful, all SQL queries can be executed directly in the notebook using the `%sql` or `%%sql` magic commands.

## Example

```python
%load_ext sql
%sql mysql+mysqlconnector://root:123456@localhost/test
```

After connecting, you can run SQL queries such as:

```sql
%%sql
SELECT * FROM employees;
```

# Introduction to SQL

**SQL (Structured Query Language)** is a standard language used to store, retrieve, update, and manage data in relational databases. It allows users to create databases, define tables, insert records, modify data, delete data, and query information efficiently.

# Popular Database Providers

## MySQL
**MySQL** is an open-source relational database management system (RDBMS) widely used for web applications. It is known for its speed, reliability, and ease of use.

## Oracle Database
**Oracle Database** is a commercial relational database management system developed by Oracle Corporation. It is designed for enterprise applications that require high performance, security, and scalability.

## Microsoft SQL Server
**Microsoft SQL Server** is a relational database management system developed by Microsoft. It is commonly used in enterprise environments and integrates well with Microsoft technologies.

## PostgreSQL
**PostgreSQL** is an open-source object-relational database management system known for its advanced features, extensibility, and standards compliance.

## SQLite
**SQLite** is a lightweight, serverless relational database that stores the entire database in a single file. It is commonly used in mobile and embedded applications.

# Difference Between Popular Database Providers

| Feature | MySQL | Oracle Database | Microsoft SQL Server | PostgreSQL | SQLite |
|---------|-------|-----------------|----------------------|------------|---------|
| License | Open Source | Commercial | Commercial | Open Source | Open Source |
| Best For | Web Applications | Enterprise Applications | Microsoft Ecosystem | Advanced Applications | Mobile & Embedded Applications |
| Scalability | High | Very High | High | Very High | Low |
| Performance | Fast | Excellent | Excellent | Excellent | Good |
| Platform | Cross-Platform | Cross-Platform | Windows/Linux | Cross-Platform | Cross-Platform |
| Cost | Free (Community Edition) | Paid | Paid | Free | Free |

# Basic SQL Syntax

Before jumping into the major topics, we will first learn the basic SQL syntax required to perform **CRUD operations**.

**CRUD** stands for:

- **C** – Create
- **R** – Read
- **U** – Update
- **D** – Delete

These operations form the foundation of working with relational databases.

# Create Database

A **database** is created using the `CREATE DATABASE` statement.

## Syntax

```sql
CREATE DATABASE database_name;
```

## Database Naming Convention

The database name should be **meaningful** and follow a proper naming convention.

**Example:**

- `mk_de_test`
  - `mk` → Company Name
  - `de` → Data Engineering Team
  - `test` → Sub-team or Project Name

## More Examples

- `abc_hr_payroll`
  - `abc` → Company Name
  - `hr` → Human Resources Team
  - `payroll` → Project Name

- `xyz_sales_analytics`
  - `xyz` → Company Name
  - `sales` → Sales Team
  - `analytics` → Project Name

- `tech_fin_reporting`
  - `tech` → Company Name
  - `fin` → Finance Team
  - `reporting` → Project Name

- `acme_dev_inventory`
  - `acme` → Company Name
  - `dev` → Development Team
  - `inventory` → Project Name


````python
%%sql

create database db_test;
````

# 1. SHOW DATABASES

## Syntax

```sql

```

## Example

```sql
SHOW DATABASES;
```


# 2. USE DATABASE

**Definition:**

The `USE DATABASE` command is used to select/use a specific database database.

## Syntax

```sql
USE database_name;
```

## Example

```sql
USE mk_de_test;
```

---

````python
%%sql
USE test;
````


# 3. DROP DATABASE

**Definition:**

The `DROP DATABASE` command is used to permanently delete an existing database along with all its tables and data.

## Syntax

```sql
DROP DATABASE database_name;
```

## Example

```sql
DROP DATABASE mk_de_test;
```

---


````python
%%sql
drop database tp_de_test;
````

# 4. CREATE TABLE

**Definition:**

The `CREATE TABLE` command is used to create a new table by defining its columns and data types.

## Syntax

```sql
CREATE TABLE table_name (
    column1 datatype,
    column2 datatype,
    column3 datatype
);
```
## Example

```sql
CREATE TABLE employees (
    emp_id INT,
    emp_name VARCHAR(50),
    salary DECIMAL(10,2)
);
```

---

````python
%%sql
create table employees_tp(
    id int,
    name varchar(100)
);
````

# 5. INSERT INTO

**Definition:**

The `INSERT INTO` command is used to insert new records into a table.

## Syntax

```sql
INSERT INTO table_name (column1, column2, ...)
VALUES (value1, value2, ...);
```

## Example

```sql
INSERT INTO employees (emp_id, emp_name, salary)
VALUES (1, 'Karthik', 50000);
```

---

# 6. SELECT

**Definition:**

The `SELECT` command is used to retrieve data from a table.

## Syntax

```sql
SELECT * FROM table_name;
```

## Example

```sql
SELECT * FROM employees;
```

---

````python
%%sql
SELECT * FROM employees_tp;
````

# 7. UPDATE

**Definition:**

The `UPDATE` command is used to modify existing records in a table.

## Syntax

```sql
UPDATE table_name
SET column_name = value
WHERE condition;
```

## Example

```sql
UPDATE employees
SET salary = 60000
WHERE emp_id = 1;
```

---


````python
%%sql

UPDATE employees_tp
SET name = "Gopal S"
where id=2;
````


# 8. DELETE

**Definition:**

The `DELETE` command is used to remove one or more records from a table.

## Syntax

```sql
DELETE FROM table_name
WHERE condition;
```

## Example

```sql
DELETE FROM employees
WHERE emp_id = 1;
```

---


````python
%%sql
DELETE FROM employees_tp
WHERE id = 1;
````


# 9. TRUNCATE

**Definition:**

The `TRUNCATE` command removes all records from a table while keeping the table structure intact.

## Syntax

```sql
TRUNCATE TABLE table_name;
```

## Example

```sql
TRUNCATE TABLE employees;
```

---



````python
%%sql

TRUNCATE TABLE employees_tp;
````


# 10. DROP TABLE

**Definition:**

The `DROP TABLE` command permanently deletes a table along with all its data and structure.

## Syntax

```sql
DROP TABLE table_name;
```

## Example

```sql
DROP TABLE employees;
```

````python
%%sql
DROP TABLE employees_tp;
````
