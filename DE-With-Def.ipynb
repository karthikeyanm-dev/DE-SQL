# Jupyter SQL Magic Commands

Jupyter Notebook provides **SQL Magic Commands** through the `ipython-sql` extension. These commands allow you to execute SQL queries directly inside notebook cells.

## Load the SQL Extension

```python
%load_ext sql
```

- Loads the **ipython-sql** extension into the current Jupyter Notebook session.
- This command only needs to be executed **once** after opening the notebook.

## Connect to a MySQL Database

```python
%sql mysql+mysqlconnector://root:123456@localhost/test
```

This command establishes a connection to the MySQL database.

### Connection String Breakdown

| Part | Description |
|------|-------------|
| `mysql+mysqlconnector` | Database dialect and Python driver used to connect to MySQL. |
| `root` | MySQL username. |
| `123456` | Password for the MySQL user. |
| `localhost` | Database server address. |
| `test` | Database name to connect to. |

Once the connection is successful, all SQL queries can be executed directly in the notebook using the `%sql` or `%%sql` magic commands.

## Example

```python
%load_ext sql
%sql mysql+mysqlconnector://root:123456@localhost/test
```

After connecting, you can run SQL queries such as:

```sql
%%sql
SELECT * FROM employees;
```

# Introduction to SQL

**SQL (Structured Query Language)** is a standard language used to store, retrieve, update, and manage data in relational databases. It allows users to create databases, define tables, insert records, modify data, delete data, and query information efficiently.

# Popular Database Providers

## MySQL
**MySQL** is an open-source relational database management system (RDBMS) widely used for web applications. It is known for its speed, reliability, and ease of use.

## Oracle Database
**Oracle Database** is a commercial relational database management system developed by Oracle Corporation. It is designed for enterprise applications that require high performance, security, and scalability.

## Microsoft SQL Server
**Microsoft SQL Server** is a relational database management system developed by Microsoft. It is commonly used in enterprise environments and integrates well with Microsoft technologies.

## PostgreSQL
**PostgreSQL** is an open-source object-relational database management system known for its advanced features, extensibility, and standards compliance.

## SQLite
**SQLite** is a lightweight, serverless relational database that stores the entire database in a single file. It is commonly used in mobile and embedded applications.

# Difference Between Popular Database Providers

# 2. CTAS (Create Table As Select)

**Definition:**

**CTAS (Create Table As Select)** is used to create a new table from the result of a `SELECT` query. The new table will contain the data returned by the query.

Suppose we already have a table named **`employees_day`** with **50 records**. We can use this table to create another table for experimentation without affecting the original data.

## Syntax

```sql
CREATE TABLE new_table_name AS
SELECT column_name(s)
FROM existing_table
WHERE condition;
```

## Example 1: Copy All Records

```sql
CREATE TABLE employees_backup AS
SELECT *
FROM employees_day;
```

## Example 2: Copy Selected Columns

```sql
CREATE TABLE employee_details AS
SELECT emp_id, emp_name, department
FROM employees_day;
```

## Example 3: Copy the Selected Columns or ALL with the specific Condition [WHERE]
```sql
CREATE TABLE employee_details_50K_salary AS
SELECT emp_id, emp_name, department
FROM employees_day where salary >= 50000;
```

## Verify the New Table

```sql
SELECT * FROM employees_backup;
```



````python
%%sql
-- Copy ALl Datas
CREATE TABLE employees_backup AS 
SELECT * 
FROM employees_day;
````

````python
%%sql
-- Just Viewing the Data 
select * from employees_backup
limit 10;
````

````python
%%sql
-- Now Copying the Specific Data

create table employees_details as 
select emp_id, emp_name
from employees_day;
````

````python
%%sql
select * from employees_details
limit 10;
````

````python
%%sql
-- Example 3
CREATE TABLE employee_details_50K_salary AS
SELECT emp_id, emp_name, department
FROM employees_day where salary >= 50000;
````

````python
%%sql
SELECT * from employee_details_50K_salary;
````

# 3. Temporary Table

**Definition:**

A **Temporary Table** is used to store data temporarily. The table exists only for the duration of the current database session and is automatically removed when the session ends.

For example, if you create a temporary table in **Jupyter Notebook**, it will only be available in that Jupyter SQL session. It will **not** be visible in another session, such as **MySQL Workbench**.

## Syntax

```sql
CREATE TEMPORARY TABLE table_name (
    column1 datatype,
    column2 datatype,
    column3 datatype
);
```

## Example

```sql
CREATE TEMPORARY TABLE temp_employees (
    emp_id INT,
    emp_name VARCHAR(100),
    department VARCHAR(50)
);
```

A temporary table can also be created using **CTAS (Create Table As Select)**.

# 4. Common Table Expression (CTE)

**Definition:**

A **Common Table Expression (CTE)** is a **temporary result set** that is defined within a SQL query and can be referenced like a **virtual table**. It exists only during the execution of that query. CTEs make complex SQL queries **easier to read, write, and maintain** by breaking them into smaller, logical parts.

---

## Syntax

```sql
WITH cte_name AS (
    SELECT column1, column2, ...
    FROM table_name
    WHERE condition
)
SELECT *
FROM cte_name;
```

---

## Execution Flow

When a CTE is executed, SQL processes it in the following order:

1. The query inside the `WITH` clause executes first.
2. The result of that query is stored as a **temporary result set** (virtual table).
3. The main query uses the CTE just like a normal table.
4. Once the query execution is completed, the CTE is automatically removed.

---

## Example

```sql
WITH high_salary_employees AS (
    SELECT *
    FROM employees_day
    WHERE salary > 50000
)

SELECT *
FROM high_salary_employees;
```

In the above example:

- The query inside the `WITH` clause executes first.
- The filtered records are stored in the temporary result set named `high_salary_employees`.
- The `SELECT` statement retrieves data from the CTE.
- After the query finishes, the CTE no longer exists.



## Example: Using CTE with CTAS

A **CTE** can also be combined with **CTAS (Create Table As Select)** to create a new table from the CTE result.

```sql
%%sql
CREATE TABLE bonus_table AS
WITH bonus_cte AS (
    SELECT
        emp_id,
        emp_name,
        department,
        salary,
        salary * 0.30 AS bonus_salary
    FROM employees_day
)
SELECT *
FROM bonus_cte;
```

### Execution Flow

1. The query inside `bonus_cte` executes first.
2. A temporary result set is created with the calculated `bonus_salary` column.
3. The `SELECT * FROM bonus_cte` statement reads the data from the CTE.
4. `CREATE TABLE bonus_table AS` stores the final result permanently in the `bonus_table`.
5. After the execution is completed, the CTE `bonus_cte` is automatically removed, while `bonus_table` remains in the database.

````python
%%sql
-- Example  1 : Create a CTE and store and select the data 
WITH bonus_cte AS (
    SELECT emp_name, salary * 0.2 as bonus_amount
    from employees_day
)
select * from bonus_cte;
````

````python
%%sql
CREATE TABLE bonus_table as 
with bonus_cte as (
    SELECT emp_id,emp_name,department,salary,salary * 0.3 as bonus_salary
    FROM employees_day
)
select * from bonus_cte;
````
