# Introduction to SQL

**SQL (Structured Query Language)** is a standard language used to store, retrieve, update, and manage data in relational databases. It allows users to create databases, define tables, insert records, modify data, delete data, and query information efficiently.

# Popular Database Providers

## MySQL
**MySQL** is an open-source relational database management system (RDBMS) widely used for web applications. It is known for its speed, reliability, and ease of use.

## Oracle Database
**Oracle Database** is a commercial relational database management system developed by Oracle Corporation. It is designed for enterprise applications that require high performance, security, and scalability.

## Microsoft SQL Server
**Microsoft SQL Server** is a relational database management system developed by Microsoft. It is commonly used in enterprise environments and integrates well with Microsoft technologies.

## PostgreSQL
**PostgreSQL** is an open-source object-relational database management system known for its advanced features, extensibility, and standards compliance.

## SQLite
**SQLite** is a lightweight, serverless relational database that stores the entire database in a single file. It is commonly used in mobile and embedded applications.

# 4. Common Table Expression (CTE)

**Definition:**

A **Common Table Expression (CTE)** is a **temporary result set** that is defined within a SQL query and can be referenced like a **virtual table**. It exists only during the execution of that query. CTEs make complex SQL queries **easier to read, write, and maintain** by breaking them into smaller, logical parts.

---

## Syntax

```sql
WITH cte_name AS (
    SELECT column1, column2, ...
    FROM table_name
    WHERE condition
)
SELECT *
FROM cte_name;
```

---

## Execution Flow

When a CTE is executed, SQL processes it in the following order:

1. The query inside the `WITH` clause executes first.
2. The result of that query is stored as a **temporary result set** (virtual table).
3. The main query uses the CTE just like a normal table.
4. Once the query execution is completed, the CTE is automatically removed.

---

## Example

```sql
WITH high_salary_employees AS (
    SELECT *
    FROM employees_day
    WHERE salary > 50000
)

SELECT *
FROM high_salary_employees;
```

In the above example:

- The query inside the `WITH` clause executes first.
- The filtered records are stored in the temporary result set named `high_salary_employees`.
- The `SELECT` statement retrieves data from the CTE.
- After the query finishes, the CTE no longer exists.



## Example: Using CTE with CTAS

A **CTE** can also be combined with **CTAS (Create Table As Select)** to create a new table from the CTE result.

```sql
%%sql
CREATE TABLE bonus_table AS
WITH bonus_cte AS (
    SELECT
        emp_id,
        emp_name,
        department,
        salary,
        salary * 0.30 AS bonus_salary
    FROM employees_day
)
SELECT *
FROM bonus_cte;
```

### Execution Flow

1. The query inside `bonus_cte` executes first.
2. A temporary result set is created with the calculated `bonus_salary` column.
3. The `SELECT * FROM bonus_cte` statement reads the data from the CTE.
4. `CREATE TABLE bonus_table AS` stores the final result permanently in the `bonus_table`.
5. After the execution is completed, the CTE `bonus_cte` is automatically removed, while `bonus_table` remains in the database.

````python
%%sql
-- Example  1 : Create a CTE and store and select the data 
WITH bonus_cte AS (
    SELECT emp_name, salary * 0.2 as bonus_amount
    from employees_day
)
select * from bonus_cte;
````

````python
%%sql
CREATE TABLE bonus_table as 
with bonus_cte as (
    SELECT emp_id,emp_name,department,salary,salary * 0.3 as bonus_salary
    FROM employees_day
)
select * from bonus_cte;
````

# SQL Query Execution Order

Although we write SQL queries in one order, the **SQL engine executes them in a different order**. Understanding the execution order helps in writing efficient and error-free queries.

---

## Example Query

```sql
SELECT department, AVG(salary) AS avg_salary
FROM employees_day
WHERE salary > 30000
GROUP BY department
HAVING AVG(salary) > 50000
ORDER BY avg_salary DESC
LIMIT 5;
```

---

# Actual Execution Order

## 1. FROM

**Purpose:**

Identifies the table(s) from which the data should be retrieved.

```sql
FROM employees_day
```

At this stage, SQL reads all the records from the `employees_day` table.

---

## 2. WHERE

**Purpose:**

Filters individual rows before any grouping or aggregation takes place.

```sql
WHERE salary > 30000
```

Only employees with a salary greater than **30,000** are considered for the next steps.

---

## 3. GROUP BY

**Purpose:**

Groups rows that have the same value in the specified column.

```sql
GROUP BY department
```

All employees are grouped according to their department.

---

## 4. HAVING

**Purpose:**

Filters the grouped data after the aggregation is performed.

```sql
HAVING AVG(salary) > 50000
```

Only departments with an average salary greater than **50,000** are retained.

---

## 5. SELECT

**Purpose:**

Chooses the columns and expressions that will appear in the final output.

```sql
SELECT department,
       AVG(salary) AS avg_salary
```

The average salary is calculated for each remaining department.

---

## 6. ORDER BY

**Purpose:**

Sorts the final result set.

```sql
ORDER BY avg_salary DESC
```

The departments are sorted in descending order of average salary.

---

## 7. LIMIT

**Purpose:**

Restricts the number of rows returned.

```sql
LIMIT 5;
```

Only the first **5** rows from the sorted result are displayed.

---

# Complete Execution Flow

```
FROM
   ↓
WHERE
   ↓
GROUP BY
   ↓
HAVING
   ↓
SELECT
   ↓
ORDER BY
   ↓
LIMIT
```

---

# Easy Way to Remember

| Execution Order | Clause | Purpose |
|----------------:|---------|---------|
| 1 | `FROM` | Reads the table(s). |
| 2 | `WHERE` | Filters rows. |
| 3 | `GROUP BY` | Creates groups. |
| 4 | `HAVING` | Filters groups. |
| 5 | `SELECT` | Chooses columns and expressions. |
| 6 | `ORDER BY` | Sorts the result. |
| 7 | `LIMIT` | Restricts the number of rows returned. |

# Constraints

**Definition:**

**Constraints** are rules applied to table columns to enforce data integrity, accuracy, and consistency in a database. They help ensure that only valid data is stored in the table.

Some of the commonly used SQL constraints are:

- Primary Key
- Foreign Key
- Unique
- Not Null
- Check
- Default

---

# 1. PRIMARY KEY

**Definition:**

A **Primary Key** is a column (or a combination of columns) that uniquely identifies each record in a table. A primary key **cannot contain `NULL` values** and **duplicate values are not allowed**.

## Characteristics

- Uniquely identifies each row.
- Does not allow duplicate values.
- Does not allow `NULL` values.
- A table can have **only one Primary Key**.
- A Primary Key can consist of one or more columns (Composite Primary Key).

---

## Syntax

```sql
CREATE TABLE table_name (
    column_name datatype PRIMARY KEY,
    column2 datatype,
    column3 datatype
);
```

---

## Example


# 2. FOREIGN KEY

**Definition:**

A **Foreign Key** is a column (or a combination of columns) in one table that refers to the **Primary Key** of another table. It is used to establish a relationship between two tables and maintain **referential integrity**.

A Foreign Key allows duplicate values and can contain `NULL` values (unless restricted by a `NOT NULL` constraint).

---

## Characteristics

- Creates a relationship between two tables.
- References the Primary Key of another table.
- Maintains referential integrity.
- Duplicate values are allowed.
- `NULL` values are allowed (unless specified otherwise).

---

## Syntax

```sql
CREATE TABLE child_table (
    column1 datatype,
    foreign_key_column datatype,
    FOREIGN KEY (foreign_key_column)
    REFERENCES parent_table(primary_key_column)
);
```

---

## Example

### Parent Table

```sql
CREATE TABLE departments (
    department_id INT PRIMARY KEY,
    department_name VARCHAR(100)
);
```

### Child Table

```sql
CREATE TABLE employees (
    emp_id INT PRIMARY KEY,
    emp_name VARCHAR(100),
    department_id INT,
    FOREIGN KEY (department_id)
    REFERENCES departments(department_id)
);
```

In this example:

- `department_id` in the **departments** table is the **Primary Key**.
- `department_id` in the **employees** table is the **Foreign Key**.
- Every `department_id` in the `employees` table must exist in the `departments` table.

---

## Adding a Foreign Key Using ALTER

### Syntax

```sql
ALTER TABLE table_name
ADD CONSTRAINT constraint_name
FOREIGN KEY (column_name)
REFERENCES parent_table(primary_key_column);
```

### Example

```sql
ALTER TABLE employees
ADD CONSTRAINT fk_department
FOREIGN KEY (department_id)
REFERENCES departments(department_id);

```

# Real-Life Example of a FOREIGN KEY

Consider a **Cab Booking Application** such as Uber or Ola.

- A **Driver** can complete multiple rides.
- Each **Ride** belongs to one Driver.
- This relationship is established using a **Foreign Key**.

---

## Parent Table - Drivers

```sql
CREATE TABLE drivers (
    driver_id INT PRIMARY KEY,
    driver_name VARCHAR(100) NOT NULL,
    license_number VARCHAR(30) UNIQUE
);
```

---

## Child Table - Rides

```sql
CREATE TABLE rides (
    ride_id INT PRIMARY KEY,
    pickup_location VARCHAR(100),
    dropoff_location VARCHAR(100),
    ride_date DATE,
    fare DECIMAL(10,2),
    driver_id INT,

    FOREIGN KEY (driver_id)
    REFERENCES drivers(driver_id)
    ON DELETE CASCADE
);
```

---

## Relationship

```
Drivers (Parent)
------------------------
driver_id (PK)
driver_name
license_number
      │
      │ One Driver
      │
      ▼
Rides (Child)
------------------------
ride_id (PK)
pickup_location
dropoff_location
ride_date
fare
driver_id (FK)
```

One **Driver** can have **multiple Rides**, but each **Ride** belongs to only one Driver.

---

# What Does `ON DELETE CASCADE` Do?

Suppose the following data exists.

### Drivers

| driver_id | driver_name |
|-----------|-------------|
| 101 | Rahul |

### Rides

| ride_id | driver_id |
|---------|-----------|
| 1 | 101 |
| 2 | 101 |
| 3 | 101 |

If we execute:

```sql
DELETE FROM drivers
WHERE driver_id = 101;
```

Because the Foreign Key uses **`ON DELETE CASCADE`**, MySQL automatically deletes:

- Driver `101`
- Ride `1`
- Ride `2`
- Ride `3`

This prevents orphan records from remaining in the `rides` table.

---

# What Happens Without `ON DELETE CASCADE`?

If the Foreign Key is created **without** `ON DELETE CASCADE`:

```sql
FOREIGN KEY (driver_id)
REFERENCES drivers(driver_id);
```

Then executing:

```sql
DELETE FROM drivers
WHERE driver_id = 101;
```

will result in an error similar to:

```
Cannot delete or update a parent row:
a foreign key constraint fails.
```

This happens because ride records still reference that driver.

---

# Is Physically Deleting Ride Data Always a Good Idea?

Not always.

Imagine a ride-booking company.

A driver leaves the company, but the completed ride history is still important for:

- Revenue reports
- Customer complaints
- Tax records
- Audits
- Business analytics

If the driver and all rides are deleted using **CASCADE**, the company loses valuable historical data.

---

# Soft Delete

Instead of deleting records permanently, many companies use a **Soft Delete**.

A new column is added to indicate whether the record is active or deleted.

---

## Add the Soft Delete Column

```sql
ALTER TABLE drivers
ADD COLUMN is_deleted BOOLEAN DEFAULT FALSE;
```

---

## Soft Delete a Driver

Instead of deleting the row:

```sql
DELETE FROM drivers
WHERE driver_id = 101;
```

we update the flag:

```sql
UPDATE drivers
SET is_deleted = TRUE
WHERE driver_id = 101;
```

The record still exists in the table, but it is marked as deleted.

---

## Fetch Only Active Drivers

```sql
SELECT *
FROM drivers
WHERE is_deleted = FALSE;
```

---

# Advantages of Soft Delete

- Data is not permanently lost.
- Historical ride records remain available.
- Reports and analytics remain accurate.
- Deleted records can be restored if needed.
- Supports auditing and regulatory compliance.

---

# Summary

| ON DELETE CASCADE | Soft Delete |
|-------------------|-------------|
| Permanently deletes parent and child records. | Marks records as deleted without removing them. |
| No historical data is retained. | Historical data is preserved. |
| Suitable for temporary or non-critical data. | Preferred for business-critical applications. |
| Automatic deletion of related records. | Requires filtering using the `is_deleted` flag. |

> **Industry Practice:** Most enterprise applications (banking, healthcare, e-commerce, ride-booking, ERP, etc.) prefer **Soft Delete** over permanent deletion to preserve historical data and maintain audit trails.

# Minimal Key (Conceptual)

## Definition

A **Minimal Key** is the **smallest set of attributes (columns)** that can uniquely identify every row in a table.

A key is considered **minimal** if:

- It uniquely identifies every row in the table.
- No column can be removed from the key while still maintaining uniqueness.

If removing even one column causes the key to lose its uniqueness, then it is a **Minimal Key**.

---

## Characteristics

- Uniquely identifies every row.
- Contains the minimum number of columns required for uniqueness.
- Removing any column results in the loss of uniqueness.
- Every Primary Key is a Minimal Key, but not every Minimal Key is chosen as the Primary Key.

---

# Example of a Minimal Key

Consider the following table:

| student_id | course_id | course_name |
|------------|-----------|-------------|
| 101 | C001 | SQL |
| 101 | C002 | Python |
| 102 | C001 | SQL |

Suppose the Primary Key is defined as:

```sql
PRIMARY KEY (student_id, course_id)
```

This is a **Minimal Key** because:

- `student_id` alone **cannot** uniquely identify a row.
- `course_id` alone **cannot** uniquely identify a row.
- The combination `(student_id, course_id)` uniquely identifies every record.
- Removing either column would result in duplicate records.

Therefore, `(student_id, course_id)` is a **Minimal Key**.

---

# Example of a Non-Minimal Key

Consider the following key:

```text
(student_id, course_id, course_name)
```

Although this combination can uniquely identify a row, it is **not a Minimal Key**.

Why?

Because:

- `(student_id, course_id)` already uniquely identifies every row.
- `course_name` is unnecessary.
- Removing `course_name` still maintains uniqueness.

Therefore,

```text
(student_id, course_id, course_name)
```

is a **Non-Minimal Key** because it contains an extra column.

---

# Comparison

| Minimal Key | Non-Minimal Key |
|--------------|-----------------|
| Smallest set of columns required for uniqueness. | Contains one or more unnecessary columns. |
| No column can be removed. | One or more columns can be removed without affecting uniqueness. |
| Efficient and optimized. | Redundant and less efficient. |

---

## Summary

- **Minimal Key** = Smallest possible unique identifier.
- Every column in the key is necessary.
- Removing any column breaks uniqueness.
- Example of a Minimal Key:
  ```text
  (student_id, course_id)
  ```
- Example of a Non-Minimal Key:
  ```text
  (student_id, course_id, course_name)
  ```

# Natural Key (Conceptual)

## Definition

A **Natural Key** is a column (or a combination of columns) that **naturally exists in the real world** and can uniquely identify each record in a table. It is derived from real-world business data rather than being generated by the database.

---

## Characteristics

- Naturally exists in the business domain.
- Uniquely identifies each record.
- Does not need to be generated by the database.
- Usually has a real-world meaning.
- Can be used as a Primary Key if it is stable and guaranteed to be unique.

---

# Real-World Examples

| Entity | Natural Key |
|---------|-------------|
| Citizen | Aadhaar Number |
| Employee | Employee ID (assigned by the organization) |
| Student | Roll Number |
| Vehicle | Vehicle Registration Number |
| Passport | Passport Number |
| Book | ISBN Number |
| Email Account | Email Address (if unique) |

---

# Example

Consider a **Citizens** table.

| aadhaar_number | citizen_name | city |
|----------------|--------------|------|
| 1234-5678-9012 | Rahul | Chennai |
| 9876-5432-1001 | Priya | Bangalore |

Here, the **Aadhaar Number** is a **Natural Key** because:

- It naturally exists in the real world.
- It uniquely identifies each citizen.
- No two citizens can have the same Aadhaar Number.

---

# Natural Key vs Surrogate Key

| Natural Key | Surrogate Key |
|--------------|---------------|
| Derived from real-world data. | Generated by the database. |
| Has business meaning. | Has no business meaning. |
| Example: Aadhaar Number, Passport Number. | Example: `customer_id`, `emp_id` (Auto Increment). |
| May change in some cases. | Usually never changes. |

---

## Summary

- A **Natural Key** is a real-world unique identifier.
- It has business meaning.
- It naturally exists in the data.
- Examples include:
  - Aadhaar Number
  - Passport Number
  - Vehicle Registration Number
  - ISBN Number
  - Employee ID

# Surrogate Key (Conceptual)

## Definition

A **Surrogate Key** is a **system-generated unique identifier** used to uniquely identify each record in a table. It has **no business meaning** and is typically generated automatically by the database using features such as **AUTO_INCREMENT**, **IDENTITY**, or **SEQUENCE**.

Unlike a Natural Key, a Surrogate Key remains stable even if the business data changes.

---

## Characteristics

- System-generated.
- Uniquely identifies each record.
- Has no real-world or business meaning.
- Usually an integer value.
- Does not change even if business data changes.
- Commonly used as the Primary Key in databases.

---

# Example

```sql
CREATE TABLE customers (
    customer_key INT AUTO_INCREMENT PRIMARY KEY,
    customer_name VARCHAR(100),
    email VARCHAR(100),
    phone_number VARCHAR(15)
);
```

Here:

- `customer_key` is the **Surrogate Key**.
- It is automatically generated by the database.
- It has no business meaning.
- Each new customer receives a unique value.

---

# Example Data

| customer_key | customer_name | email |
|--------------|---------------|------------------------|
| 1 | Rahul | rahul@gmail.com |
| 2 | Priya | priya@gmail.com |
| 3 | Arun | arun@gmail.com |

The values **1, 2, 3...** are generated automatically using `AUTO_INCREMENT`.

---

# Real-World ETL Example

In a **Data Warehouse**, customer information may change over time (for example, a customer changes their address or phone number). Instead of updating the existing record, a **new version** of the customer record is inserted.

Each version receives a **new Surrogate Key**, allowing the warehouse to preserve historical data.

| customer_key | customer_id | customer_name | city |
|--------------|-------------|---------------|------|
| 101 | C001 | Rahul | Chennai |
| 102 | C001 | Rahul | Bangalore |

In this example:

- `customer_id (C001)` is the business (Natural) Key.
- `customer_key` is the **Surrogate Key**.
- Although the customer is the same, each version of the record has a different surrogate key.

This is why a Surrogate Key is often described as being **unique for each version of a customer record** in a data warehouse.

---

# Natural Key vs Surrogate Key

| Natural Key | Surrogate Key |
|--------------|---------------|
| Derived from business data. | Generated by the database. |
| Has business meaning. | Has no businessm meaning. |
| May change over time. | Never changes. |
| Example: Aadhaar Number, Customer ID. | Example: AUTO_INCREMENT, IDENTITY, SEQUENCE. |

---

## Summary

- A **Surrogate Key** is a system-generated unique identifier.
- It is commonly created using `AUTO_INCREMENT`.
- It has no business meaning.
- It is stable and does not change.
- In ETL and Data Warehousing, it is commonly used to uniquely identify **each version of a record**, making it ideal for maintaining historical data.

# Candidate Key (Conceptual)

## Definition

A **Candidate Key** is a column or a combination of columns that can **uniquely identify every row** in a table. A table can have **one or more Candidate Keys**, and one of them is selected as the **Primary Key**.

A Candidate Key must be **unique** and **minimal**, meaning it should not contain any unnecessary columns.

---

## Characteristics

- Uniquely identifies every row.
- Contains the minimum number of columns required for uniqueness.
- A table can have multiple Candidate Keys.
- One Candidate Key is chosen as the Primary Key.
- A Candidate Key cannot contain unnecessary (optional) columns.

---

# Relationship with Primary Key

- Every **Primary Key** is a **Candidate Key**.
- Every **Composite Primary Key** is also a **Candidate Key**.
- **Not every Candidate Key becomes the Primary Key.**

---

# Example 1

Consider the following table:

| student_id | email | phone_number | student_name |
|------------|-------|--------------|--------------|
| 101 | a@gmail.com | 9876543210 | Rahul |
| 102 | b@gmail.com | 9876543211 | Priya |

Possible Candidate Keys:

- `student_id`
- `email`
- `phone_number`

Each of these columns can uniquely identify a student.

Suppose we choose:

```sql
PRIMARY KEY (student_id)
```

Then:

- `student_id` → Primary Key (also a Candidate Key)
- `email` → Candidate Key
- `phone_number` → Candidate Key

---

# Example 2

Suppose the Primary Key is:

```sql
PRIMARY KEY (student_id, course_id)
```

The combination:

```text
(student_id, course_id)
```

is a **Candidate Key** because:

- It uniquely identifies each record.
- Neither column can be removed while maintaining uniqueness.

---

# Example of a Non-Candidate Key

Consider the following combination:

```text
(student_id, course_id, course_name)
```

This is **not** a Candidate Key because:

- `(student_id, course_id)` already uniquely identifies the row.
- `course_name` is an extra (optional) column.
- Removing `course_name` does not affect uniqueness.

Therefore, it is **not minimal**, so it cannot be a Candidate Key.

---

# Summary

| Candidate Key | Non-Candidate Key |
|----------------|-------------------|
| Uniquely identifies every row. | May uniquely identify a row but contains unnecessary columns. |
| Minimal set of columns. | Contains extra columns. |
| Can be selected as the Primary Key. | Cannot be selected as a Candidate Key because it is not minimal. |

---

## Key Points

- A table can have **multiple Candidate Keys**.
- One Candidate Key is chosen as the **Primary Key**.
- Every Primary Key is a Candidate Key.
- Every Composite Primary Key is a Candidate Key.
- Candidate Keys must be **unique** and **minimal**.
- Example of a Candidate Key:
  ```text
  student_id
  ```

- Example of a Non-Candidate Key:
  ```text
  (student_id, course_id, course_name)
  ```

# Super Key (Conceptual)

## Definition

A **Super Key** is a column or a combination of columns that can **uniquely identify every row** in a table.

A Super Key may contain **extra columns** that are not required for uniqueness. As long as the combination uniquely identifies each row, it is considered a Super Key.

---

## Characteristics

- Uniquely identifies every row.
- May contain one or more unnecessary columns.
- A table can have multiple Super Keys.
- Every Candidate Key is a Super Key.
- Every Primary Key is a Super Key.
- Not every Super Key is a Candidate Key.

---

# Relationship Between Keys

- **All Candidate Keys are Super Keys.**
- **All Primary Keys are Super Keys.**
- **Not all Super Keys are Candidate Keys.**

The reason is that a Candidate Key must be **minimal**, whereas a Super Key **may include extra columns**.

---

# Example

Consider the following table:

| student_id | course_id | course_name |
|------------|-----------|-------------|
| 101 | C001 | SQL |
| 101 | C002 | Python |
| 102 | C001 | SQL |

Suppose the Primary Key is:

```sql
PRIMARY KEY (student_id, course_id)
```

This is a **Candidate Key** as well as a **Super Key** because it uniquely identifies every row.

---

## Examples of Super Keys

The following combinations are all **Super Keys** because they uniquely identify each row.

```text
(student_id, course_id)
```

```text
(student_id, course_id, course_name)
```

```text
(student_id, course_id, course_name, student_name)
```

Although the additional columns (`course_name`, `student_name`) are **not required** for uniqueness, the combinations still uniquely identify each row. Therefore, they are **Super Keys**.

---

# Example of a Candidate Key

```text
(student_id, course_id)
```

This is a **Candidate Key** because:

- It uniquely identifies every row.
- It contains no unnecessary columns.

---

# Example of a Super Key but NOT a Candidate Key

```text
(student_id, course_id, course_name)
```

This is a **Super Key** because it uniquely identifies every row.

However, it is **not** a Candidate Key because:

- `course_name` is an extra column.
- Removing `course_name` still maintains uniqueness.
- Therefore, it is **not minimal**.

---

# Comparison

| Candidate Key | Super Key |
|---------------|-----------|
| Must be minimal. | May contain extra columns. |
| No unnecessary columns. | Can contain unnecessary columns. |
| Every Candidate Key is a Super Key. | Not every Super Key is a Candidate Key. |
| Can be chosen as the Primary Key. | Cannot be chosen as the Primary Key if it is not minimal. |

---

## Summary

- A **Super Key** is any column or combination of columns that uniquely identifies a row.
- It **may include additional columns** that are not required for uniqueness.
- **All Primary Keys are Super Keys.**
- **All Candidate Keys are Super Keys.**
- **Not all Super Keys are Candidate Keys.**

### Examples

**Candidate Key (also a Super Key)**

```text
(student_id, course_id)
```

**Super Key (Not a Candidate Key)**

```text
(student_id, course_id, course_name)
```

# Aggregate Functions

## Definition

**Aggregate Functions** are SQL functions that perform calculations on **multiple rows of data** and return a **single result**. They are commonly used to summarize data, such as calculating totals, averages, minimum values, maximum values, and row counts.

Aggregate functions are often used with the `GROUP BY` clause to perform calculations for each group of records.

---

## Common Aggregate Functions

- `COUNT()`
- `SUM()`
- `AVG()`
- `MIN()`
- `MAX()`

# 1. COUNT()

## Definition

The `COUNT()` function is an **aggregate function** used to count the number of rows in a table or the number of non-`NULL` values in a specific column.

---

## Syntax

### Count All Rows

```sql
SELECT COUNT(*)
FROM table_name;
```

### Count Non-NULL Values in a Column

```sql
SELECT COUNT(column_name)
FROM table_name;
```

---

## Example 1: Count All Employees

```sql
SELECT COUNT(*)
FROM employees_day;
```

This query returns the **total number of rows (employees)** in the `employees_day` table.

---

## Example 2: Count Employees in a Specific Department

```sql
SELECT COUNT(*)
FROM employees_day
WHERE department = 'IT';
```

This query returns the **number of employees** who belong to the **IT** department.

---

## Example 3: Count Non-NULL Values

```sql
SELECT COUNT(email)
FROM employees_day;
```

This query counts only the rows where the `email` column is **not NULL**.

---

## Note

- `COUNT(*)` counts **all rows**, including rows that contain `NULL` values.
- `COUNT(column_name)` counts only the rows where the specified column is **not NULL**.

````python
%%sql 
SELECT COUNT(*) 
FROM employees_day;
````

````python
%%sql
SELECT COUNT(*)
FROM employees_day
WHERE department = 'IT';
````

## Syntax

# HAVING Clause

## Definition

The `HAVING` clause is used to **filter grouped data** after the `GROUP BY` operation has been performed. It is similar to the `WHERE` clause, but it works on **groups** instead of individual rows.

- `WHERE` filters **rows** before grouping.
- `HAVING` filters **groups** after grouping.

---

## Why Do We Need HAVING?

We can use the `WHERE` clause to filter rows, but **aggregate functions cannot be used in the `WHERE` clause**.

Consider the following query:

```sql
SELECT department, AVG(salary)
FROM employees_day
GROUP BY department;
```

This query groups all employees by **department** and calculates the average salary for each department.

Suppose we only want the departments whose **average salary is greater than 50,000**.

The following query **does not work**:

```sql
SELECT department, AVG(salary)
FROM employees_day
WHERE AVG(salary) > 50000
GROUP BY department;
```

**Reason:**

The `WHERE` clause executes **before** `GROUP BY`, so the average salary has not been calculated yet.

---

## Correct Solution Using HAVING

```sql
SELECT department, AVG(salary)
FROM employees_day
GROUP BY department
HAVING AVG(salary) > 50000;
```

### Output

```text
Department    AVG(salary)
-------------------------
IT            62000
Finance       57000
```

Only the departments whose average salary is greater than **50,000** are displayed.

---

## Syntax

```sql
SELECT column_name, aggregate_function(column_name)
FROM table_name
GROUP BY column_name
HAVING aggregate_function(column_name) condition;
```

---

## WHERE vs HAVING

| WHERE | HAVING |
|--------|---------|
| Filters individual rows. | Filters grouped data. |
| Executes before `GROUP BY`. | Executes after `GROUP BY`. |
| Cannot use aggregate functions. | Can use aggregate functions. |

---

## Execution Order

```text
FROM
   ↓
WHERE
   ↓
GROUP BY
   ↓
HAVING
   ↓
SELECT
   ↓
ORDER BY
   ↓
LIMIT
```

---

## Summary

- `WHERE` filters **rows**.
- `HAVING` filters **groups**.
- `HAVING` is used only after `GROUP BY`.
- Aggregate functions such as `COUNT()`, `SUM()`, `AVG()`, `MIN()`, and `MAX()` are commonly used with the `HAVING` clause.

````python
%%sql 
select department,avg(salary) avgSal
from employees_day
group by department
having avgSal > 50000;
````

# Using WHERE and HAVING Together

In many real-world scenarios, we need to use **both `WHERE` and `HAVING`** in the same query.

- **`WHERE`** filters **individual rows** before grouping.
- **`GROUP BY`** groups the remaining rows.
- **`HAVING`** filters the grouped results after the aggregation is performed.

---

## Example

```sql
SELECT department, AVG(salary)
FROM employees_day
WHERE salary > 30000
GROUP BY department
HAVING AVG(salary) > 45000;
```

---

## Execution Flow

### Step 1: WHERE


- Employees with a salary less than or equal to **30,000** are removed.
- Only the remaining rows proceed to the next step.

⬇️

### Step 2: GROUP BY

- The remaining employees are grouped based on their department.

⬇️

### Step 3: AVG()

```sql
AVG(salary)
```

- The average salary is calculated for each department.

⬇️

### Step 4: HAVING

```sql
HAVING AVG(salary) > 45000
```

- Only departments whose average salary is greater than **45,000** are returned.

---

## Visual Flow

```text
      ▼

      │
      ▼

      │

      │
      ▼
HAVING AVG(salary) > 45000
      │
      ▼
Final Result
```

---

## Key Difference

| WHERE | HAVING |
|--------|---------|
| Filters rows before grouping. | Filters groups after grouping. |
| Cannot use aggregate functions. | Can use aggregate functions. |
| Executes before `GROUP BY`. | Executes after `GROUP BY`. |

---

## Summary

- Use **`WHERE`** to filter individual rows.
- Use **`HAVING`** to filter grouped results.

```text

   ↓

   ↓

   ↓

   ↓
```

# NULL Handling

## Definition

A **NULL** value represents **unknown, missing, or unavailable data**. It does **not** mean zero (`0`), an empty string (`''`), or `FALSE`. It simply indicates that no value exists for that column.

---

## Important Points

- `NULL` means the value is **unknown**, **missing**, or **not available**.
- `NULL` is different from `0`.
- `NULL` is different from an empty string (`''`).
- Arithmetic operations involving `NULL` usually return `NULL`.
- Aggregate functions (except `COUNT(*)`) ignore `NULL` values.

---

## Why Can't We Use `=` or `!=`?

A `NULL` value is **unknown**, so SQL cannot determine whether it is equal to or not equal to another value.

Therefore, the following queries **do not work** as expected:

### ❌ Incorrect

```sql
SELECT *
FROM employees_day
WHERE department = NULL;
```

```sql
SELECT *
FROM employees_day
WHERE department != NULL;
```

These queries return **no rows** because `NULL` cannot be compared using comparison operators.

---

## Correct Way

Use `IS NULL` and `IS NOT NULL`.

### Find NULL Values

```sql
SELECT *
FROM employees_day
WHERE department IS NULL;
```

---

### Find Non-NULL Values

```sql
SELECT *
FROM employees_day
WHERE department IS NOT NULL;
```

---

## NULL Handling Functions

### IFNULL()

Returns a specified value if the expression is `NULL`.

#### Syntax

```sql
IFNULL(expression, replacement_value)
```

#### Example

```sql
SELECT
    emp_name,
    IFNULL(email, 'Not Available') AS email
FROM employees_day;
```

If the `email` is `NULL`, the query displays **'Not Available'**.

---

## Real-Time Applications

NULL handling is commonly used in:

- Replacing missing values in reports.
- Data cleaning during ETL processes.
- Displaying default values in dashboards.
- Customer profile management.
- Financial and banking applications.
- HR and payroll systems.
- Data Warehousing and Business Intelligence.
- Healthcare and insurance systems.
- CRM applications.

---

## Summary

- `NULL` represents **unknown, missing, or unavailable** data.
- `NULL` cannot be compared using `=` or `!=`.
- Use `IS NULL` to find missing values.
- Use `IS NOT NULL` to find available values.
- Use `IFNULL()` to replace `NULL` with a default value.
- Use `COALESCE()` to return the first non-NULL value from multiple expressions.

````python
%%sql
-- added a column on number and all has null value in it so we getting 50 records as total count 
select count(*) from employees_day
where number is null;
````

````python
%%sql
-- How many has null in phone number

SELECT count(*) from employees_day
where number IS NOT NULL;
````

# Subquery

## Definition

A **Subquery** is simply a **query inside another query**. The inner query (**Subquery**) executes first, and its result is passed to the outer query (**Main Query**) for further processing.

---

## Structure

```text
                Main Query
                     │
                     │ Uses the result
                     ▲
                     │
                Subquery
          (Executes First)
```

The **Subquery** always executes first, and its result is then used by the **Main Query**.

---

# Why Do We Need a Subquery?

Suppose we have an `employees_day` table, and we need to find all employees who earn **more than the average salary**.

A beginner might try to write:

```sql
SELECT *
FROM employees_day
WHERE salary > AVG(salary);
```

❌ **This query is invalid.**

**Reason:**

The `WHERE` clause executes **before** aggregate functions such as `AVG()`. Therefore, the average salary has not been calculated when the `WHERE` clause is executed.

---

# Option 1: Execute Two Queries

### Step 1: Find the Average Salary

```sql
SELECT AVG(salary)
FROM employees_day;
```

### Output

```text
45000
```

---

### Step 2: Use the Result

```sql
SELECT *
FROM employees_day
WHERE salary > 45000;
```

This approach works, but we have to execute **two separate queries** manually.

---

# Option 2: Using a Subquery

Instead of executing two queries manually, we can combine them into a single query.

```sql
SELECT *
FROM employees_day
WHERE salary > (
    SELECT AVG(salary)
    FROM employees_day
);
```

The SQL engine automatically executes the inner query first and then passes its result to the outer query.

---

# Execution Flow

Consider the following query:

```sql
SELECT *
FROM employees_day
WHERE salary > (
    SELECT AVG(salary)
    FROM employees_day
);
```

### Step 1: Execute the Subquery

```sql
SELECT AVG(salary)
FROM employees_day;
```

Suppose the result is:

```text
45000
```


### Step 2: Replace the Subquery with the Result

The query now becomes:

```sql
SELECT *
FROM employees_day
WHERE salary > 45000;
```


### Step 3: Execute the Main Query

The SQL engine returns all employees whose salary is greater than **45,000**.

---

# Execution Diagram

```text
                 Main Query

SELECT *
FROM employees_day
WHERE salary >
          │
          ▼
     Subquery Executes

SELECT AVG(salary)
FROM employees_day

          │
          ▼
      Result = 45000

          │
          ▼

Main Query Becomes

SELECT *
FROM employees_day
WHERE salary > 45000;

          │
          ▼

      Final Result
```

---

## Summary

- A **Subquery** is a query inside another query.
- The **Subquery** executes first.
- Its result is passed to the **Main Query**.
- Subqueries eliminate the need to execute multiple queries manually.
- They are commonly used with aggregate functions, filtering, joins, and data comparison.

---

## Real-Time Applications

Subqueries are commonly used for:

- Finding employees earning above the average salary.
- Comparing records with aggregate values.
- Data filtering in ETL pipelines.
- Data Warehousing.
- Business Intelligence and dashboard reporting.
- Banking and financial reporting.
- Customer and sales analytics.

````python
%%sql
SELECT
    emp_id,
    emp_name,
    department,
    salary,
    COALESCE(number, 'No Contact') AS contact_number
FROM employees_day
WHERE salary > (
    SELECT AVG(salary)
    FROM employees_day
);
````

# 3. Correlated Subquery

## Definition

A **Correlated Subquery** is a subquery that **depends on the outer query** for its values. Unlike a normal subquery, it **cannot execute independently** because it references columns from the outer query.

For **every row processed by the outer query**, the correlated subquery is executed once.

---

## Key Characteristics

- Depends on the outer query.
- Cannot execute independently.
- Executes once for **each row** of the outer query.
- Usually references columns from the outer query.
- Generally slower than a normal subquery because it executes multiple times.

---

## Syntax

```sql
SELECT column_name(s)
FROM table_name outer_table
WHERE condition operator (
    SELECT expression
    FROM table_name inner_table
    WHERE inner_table.column = outer_table.column
);
```

---

## Example

Find employees whose salary is greater than the **average salary of their own department**.

```sql
SELECT *
FROM employees_day e1
WHERE salary > (
    SELECT AVG(salary)
    FROM employees_day e2
    WHERE e1.department = e2.department
);
```

---

## How It Works

Suppose the table contains:

| emp_name | department | salary |
|----------|------------|--------|
| Rahul | IT | 70000 |
| Arun | IT | 50000 |
| Priya | HR | 40000 |
| Meena | HR | 35000 |

### Step 1

The outer query reads the first row.

```text
Rahul
Department = IT
```

The subquery becomes:

```sql
SELECT AVG(salary)
FROM employees_day
WHERE department = 'IT';
```

Output:

```text
60000
```

Outer query checks:

```text
70000 > 60000 ✔
```

Rahul is returned.

---

### Step 2

The outer query moves to the next employee.

```text
Arun
Department = IT
```

The subquery executes **again**.

```sql
SELECT AVG(salary)
FROM employees_day
WHERE department = 'IT';
```

Output:

```text
60000
```

Check:

```text
50000 > 60000 ✘
```

Arun is not returned.

The same process continues for every employee.

---

## Execution Flow

```text
Outer Query Reads First Row
            │
            ▼
Execute Correlated Subquery
            │
            ▼
Return Result
            │
            ▼
Evaluate Condition
            │
            ▼
Return Row (if condition is true)
            │
            ▼
Move to Next Row
            │
            ▼
Execute Subquery Again
```

Unlike a normal subquery, the subquery is executed **once for every row** processed by the outer query.

---

## Difference Between Normal and Correlated Subquery

| Normal Subquery | Correlated Subquery |
|-----------------|---------------------|
| Executes only once. | Executes once for each row of the outer query. |
| Can execute independently. | Depends on the outer query. |
| Faster in most cases. | Usually slower because it executes multiple times. |
| Does not reference the outer query. | References columns from the outer query. |

---

## Summary

- A **Correlated Subquery** depends on the outer query.
- It cannot execute independently.
- It executes once for every row processed by the outer query.
- It is commonly used when comparisons need to be made within groups, such as comparing an employee's salary to the average salary of their department.

---

## Real-Time Applications

Correlated Subqueries are commonly used for:

- Finding employees earning above their department's average salary.
- Comparing products within the same category.
- Customer purchase analysis.

# Views

## What is a View?

A **View** is a **virtual table** created from the result of a SQL query.

It does **not store the actual data**.
Instead, it stores only the **SQL query**, and whenever the view is accessed, SQL executes the stored query and returns the latest data.

### Simple Definition

> A **View** is a virtual table that displays data from one or more tables without storing the data itself.

---

# JOIN

## What is a JOIN?

A **JOIN** is a SQL operation that combines rows from **two or more tables** based on a related column.

The related column is usually a **Primary Key (PK)** in one table and a **Foreign Key (FK)** in another table.

### Simple Definition

> A **JOIN** is used to retrieve related data from multiple tables by matching a common column.

---

# Why Do We Need JOINs?

In relational databases, data is stored in **multiple tables** to avoid redundancy and improve data integrity.

Instead of storing everything in one table, related information is separated into different tables.

For example, in Swiggy:

### Customers

| customer_id | customer_name |
|-------------|---------------|
| 1 | Rahul |
| 2 | Priya |

### Orders

| order_id | customer_id | total_amount |
|----------|-------------|-------------:|
| 101 | 1 | 850 |
| 102 | 2 | 1200 |

The `Orders` table stores only the `customer_id`, not the customer's name.

To display:

| order_id | customer_name | total_amount |
|----------|---------------|-------------:|
| 101 | Rahul | 850 |
| 102 | Priya | 1200 |

we need a **JOIN**.

---

# Real-Life Analogy

Imagine a school.

### Students Table

| student_id | student_name |
|------------|--------------|
| 101 | Rahul |
| 102 | Priya |

### Marks Table

| student_id | subject | marks |
|------------|----------|------:|
| 101 | SQL | 95 |
| 102 | Python | 90 |

The principal asks:

> **"Show every student's name along with their marks."**

Neither table contains all the required information.

Using the common column:

```
student_id
```

we combine both tables.

This process is called a **JOIN**.

---

# General Syntax

```sql
SELECT column_list
FROM table1
JOIN table2
ON table1.common_column = table2.common_column;
```

Example:

```sql

    c.customer_name,
    o.order_id,
    o.total_amount
FROM customers c
JOIN orders o
ON c.customer_id = o.customer_id;
```

# Key Points

- JOINs eliminate the need to store duplicate data.

# INNER JOIN

## What is an INNER JOIN?

An **INNER JOIN** returns **only the rows that have matching values in both tables**.

If there is **no matching record** in either table, that row is **not included** in the result.

### Simple Definition

> An **INNER JOIN** combines two or more tables and returns **only the matching records** based on a common column.

---

# Real-Life Analogy

Imagine a college.

### Students Table

| student_id | student_name |
|------------|--------------|
| 101 | Rahul |
| 102 | Priya |
| 103 | Arjun |

### Marks Table

| student_id | subject | marks |
|------------|----------|------:|
| 101 | SQL | 95 |
| 102 | Python | 90 |
| 104 | Java | 88 |

Now the principal asks:

> **"Show students along with their marks."**

What happens?

- Rahul has marks ✅
- Priya has marks ✅
- Arjun has no marks ❌
- Student 104 has no student record ❌

Result:

| Student | Subject | Marks |
|---------|----------|------:|
| Rahul | SQL | 95 |
| Priya | Python | 90 |

Only the **matching student IDs** are returned.

This is exactly how an **INNER JOIN** works.

---

# Swiggy Example

### Customers

| customer_id | customer_name |
|-------------|---------------|
| 1 | Rahul |
| 2 | Priya |
| 3 | Arjun |

---

### Orders

| order_id | customer_id | total_amount |
|----------|-------------|-------------:|
| 101 | 1 | 850 |
| 102 | 2 | 1200 |
| 103 | 4 | 950 |

---

We need:

| order_id | customer_name | total_amount |
|----------|---------------|-------------:|
| 101 | Rahul | 850 |
| 102 | Priya | 1200 |

Notice:

- Customer 3 has no order ❌
- Order 103 belongs to customer 4, who doesn't exist ❌

Only matching records are returned.

---

# Visual Representation

```
Customers               Orders

  ○ ○ ● ●
     \ /
      ●
     / \
  ● ○ ○

Only the overlapping (matching) rows are returned.
```

Think of it as the **intersection** of two tables.

---

# Syntax

```sql
SELECT columns
FROM table1
INNER JOIN table2
ON table1.common_column = table2.common_column;
```

or simply

```sql
SELECT columns
FROM table1
JOIN table2
ON table1.common_column = table2.common_column;
```

> **Note:** `JOIN` and `INNER JOIN` are exactly the same in SQL.

---
